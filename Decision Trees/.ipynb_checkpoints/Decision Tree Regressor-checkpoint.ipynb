```python
import numpy as np
import pandas as pd
import sklearn
from sklearn import datasets
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

p = sns.color_palette
np.random.seed(0)
```

```python
# Generate a dataset and plot it
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))
plt.scatter(X, y, s=40, cmap=plt.cm.Spectral)
plt.title('Dataset')
plt.show()
```

# Decision Tree Regressor

```python
clf2 = tree.DecisionTreeRegressor(max_depth=2)
clf5 = tree.DecisionTreeRegressor(max_depth=5)
clf2 = clf2.fit(X, y)
clf5 = clf5.fit(X, y)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_2 = clf2.predict(X_test)
y_5 = clf5.predict(X_test)
plt.scatter(X, y, s=40,c="darkorange",label="data")
plt.plot(X_test, y_2, color="yellowgreen",label="max_depth=2", linewidth=2, alpha=0.8)
plt.plot(X_test, y_5, color="cornflowerblue",label="max_depth=5", linewidth=2, alpha=0.8)
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
clf = tree.DecisionTreeRegressor(max_depth=1)
clf = clf.fit(X, y)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_p = clf.predict(X_test)
plt.scatter(X, y, s=40,c="darkorange",label="data")
plt.plot(X_test, y_p, color="yellowgreen",label="max_depth=1", linewidth=2, alpha=0.8)
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

#### 懒得自己一步一步推导了。。生成一个决策树桩吧。。。

```python
min_loss = 100000
threshold = 0
for i in X:
    index = X[:,0] <= i[0]
    loss_1 = 0
    mean_1 = np.mean(y[index])
    for j in y[index]:
        loss_1 += (j - mean_1)**2
    loss_2 = 0
    if len(y[~index]!=0):
        mean_2 = np.mean(y[~index])
        for j in y[~index]:
            loss_2 += (j - mean_2)**2
    if (loss_1 + loss_2)< min_loss:
        min_loss = loss_1 + loss_2
        threshold = i
```
