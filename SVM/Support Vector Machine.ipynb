```python
import numpy as np
import pandas as pd
import sklearn
import sklearn.datasets
import sklearn.linear_model
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

p = sns.color_palette
np.random.seed(0)
```

# Generating a dataset

```python
# Generate a dataset and plot it
X, Y = sklearn.datasets.make_blobs(random_state=0, n_samples = 200,centers=2, n_features = 2, cluster_std=0.6)
plt.scatter(X[:,0], X[:,1], s=40, c=Y, cmap=plt.cm.Spectral)
plt.title('Dataset')
plt.show()
```
